```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('test_scores.csv')
df
```

```python
def gradient_desent(x,y):
    m_crri = b_crri = 0
    iterations = 50
    n = len(x)
    learning_rate = 0.00001
    for i in range(iterations):
        y_predicted = m_crri * x + b_crri
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_crri = m_crri - learning_rate * md
        b_crri = b_crri - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_crri,b_crri,cost, i))
        plt.scatter(m_crri,cost)
        # plt.plot(x,y, color = 'blue')


x = np.array(df.math)
y = np.array(df.cs)
gradient_desent(x,y)
```
