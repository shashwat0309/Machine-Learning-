```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
dff = pd.read_csv('canada_per_capita_income.csv')
dff.head(3)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(dff.year,dff.per_capita_income, color = 'red' ,marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(dff[['year']], dff.per_capita_income)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(dff.year,dff.per_capita_income, color = 'red' ,marker='+')
plt.plot(dff.year, reg.predict(dff[['year']]), color = 'blue')
```

```python
 p = reg.predict(dff[['year']])
dff['Predicted Per Capita Price'] = p
```

```python
dff.to_csv('canada_per_capita_income.csv')
```
